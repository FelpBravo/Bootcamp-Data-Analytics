```python
import numpy as np
import pandas as pd
```

```python
#Dataframe a raiz de una lista de objetos
data = {"state":["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year":[2000,2001,2002,2001,2002,2003],
        "pop":[1.5,1.7,3.6,2.4,2.9,3.2]}
frame = pd.DataFrame(data)
frame
```

```python
#Ordenar dataframe respecto a las columnas que queramos
pd.DataFrame(data, columns = ["year", "state", "pop"])
```

```python
#Ordenar dataframe y agregar una columna
pd.DataFrame(data, columns = ["year", "state", "pop", "debt"])
```

```python
#Añadir una columna con los mismos valores para todas las filas
#np.arange.. añade numeros aleatorios desde el 0 al 5
frame["debt"] = np.arange(6)
frame
```
